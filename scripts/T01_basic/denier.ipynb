```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def ComputeRadiusYarn(theta,rh):
    return np.sin(theta)*32*rh/np.pi
#
def ComputeAngle(ry,rh):
    return np.arcsin(ry*np.pi/(32*rh))*180/np.pi

def kgm2denier(lamb):
    return lamb*9e6

def ComputeLinearDensity(rho,angle,rh):
    rh = rh*1e-3 # mm -> m
    lamb =  (64/np.pi)*rho*np.cos(angle)*rh**2 # kg/m
    return kgm2denier(lamb)
```

```python
#
```

```python
# contour plot 
theta_span = np.linspace(50,90,100)*np.pi/180
rh_span = np.linspace(0.1,0.3,100)
theta,rh = np.meshgrid(theta_span,rh_span)

radius = ComputeRadiusYarn(theta,rh)*1e3
```

```python
plt.contourf(theta,rh,radius)
colorbar = plt.colorbar()
# draw contour lines where the radius is 2 
plt.contour(theta,rh,radius,levels=[2],colors='r')
#text indicating that this line corresponds to r_y = 2
plt.text(70,0.27/2,r'$r_h = 0.125 mm$',color='r')
# plot horizontal line in 0.2
plt.axhline(0.27/2,color='g',label='r_f (exp) = 0.125')
#xlabel
plt.xlabel('Theta (degrees)')
#ylabel
plt.ylabel('Radius filament (mm)')
#color bar label radius yarn 
colorbar.set_label('Radius yarn (mm)')
```
