```python
import gmsh
import math

# Iniciar la API de Gmsh
gmsh.initialize()
gmsh.model.add("helical_spring")

# Parámetros de la elipse
a = 2  # Semieje mayor
b = 1  # Semieje menor

# Parámetros de la hélice (resorte)
R = 5      # Radio del resorte
n_turns = 5  # Número de vueltas
pitch = 2  # Distancia entre vueltas (altura entre cada giro)
num_points_per_turn = 100  # Puntos por vuelta para suavidad

# Generar puntos a lo largo de la hélice
points = []
for i in range(n_turns * num_points_per_turn + 1):
    t = i / num_points_per_turn * 2 * math.pi  # Ángulo para cada punto
    x = R * math.cos(t)  # Coordenada X
    y = R * math.sin(t)  # Coordenada Y
    z = pitch * t / (2 * math.pi)  # Coordenada Z (altura)
    point = gmsh.model.geo.addPoint(x, y, z)
    points.append(point)

# Crear la curva de la hélice a partir de los puntos
helical_curve = gmsh.model.geo.addSpline(points)

# Definir los puntos de la elipse inicial
p1 = gmsh.model.geo.addPoint(a, 0, 0)
p2 = gmsh.model.geo.addPoint(0, b, 0)
p3 = gmsh.model.geo.addPoint(-a, 0, 0)
p4 = gmsh.model.geo.addPoint(0, -b, 0)
p_center = gmsh.model.geo.addPoint(0, 0, 0)

# Crear los arcos de la elipse
e1 = gmsh.model.geo.addEllipseArc(p1, p_center, p2, p2)
e2 = gmsh.model.geo.addEllipseArc(p2, p_center, p3, p3)
e3 = gmsh.model.geo.addEllipseArc(p3, p_center, p4, p4)
e4 = gmsh.model.geo.addEllipseArc(p4, p_center, p1, p1)

# Crear el Loop de la elipse y la superficie
elipse_loop = gmsh.model.geo.addCurveLoop([e1, e2, e3, e4])
elipse_surface = gmsh.model.geo.addPlaneSurface([elipse_loop])

# Refinar la malla de la elipse
gmsh.model.mesh.setSize([p1, p2, p3, p4], 0.1)

# Sincronizar la geometría
gmsh.model.geo.synchronize()

# Generar la malla 2D de la elipse antes de la extrusión
gmsh.model.mesh.generate(2)

# Barrer la elipse a lo largo de la hélice para formar el resorte, recombinando caras
gmsh.model.geo.extrude([(2, elipse_surface)], 0, 0, 0, [helical_curve], recombine=True)

# Sincronizar la geometría después de la extrusión
gmsh.model.geo.synchronize()

# Generar la malla 3D del resorte
gmsh.model.mesh.generate(3)

# Guardar la geometría en un archivo
gmsh.write("helical_spring.inp")

# Abrir la interfaz gráfica de Gmsh para visualizar
gmsh.fltk.run()

# Finalizar Gmsh
gmsh.finalize()

```
