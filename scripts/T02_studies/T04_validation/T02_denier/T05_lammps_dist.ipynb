```python
import glob

import numpy as np
from model.lammps.conse_dist import conse_dist
```

```python
lmp_paths = glob.glob("output/*/*/lammps/")
results = []
for lmp_path in lmp_paths:
    r = conse_dist(lmp_path+"data.csv")
    results.append(r)
```

```python
for r in results:
    print(r["percent"])
```

```python
from matplotlib import pyplot as plt
```

```python
# sort by theta
results = sorted(results, key=lambda k: k['vars_lmp']["theta"])
```

```python
colors = plt.cm.jet(np.linspace(0,1,len(results)))
for i,r in enumerate(results):
    plt.plot(r["consecutive_dist"],label=r["vars_lmp"]["theta"]*180/np.pi,color=colors[i])
plt.legend()
```

```python
theta = [r["vars_lmp"]["theta"]*180/np.pi for r in results]
percent = [r["percent"] for r in results]
r0_factor = [r["vars_lmp"]["r0_factor"] for r in results]
```
